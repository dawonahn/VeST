```python
import pandas as pd
import numpy as np
import csv
from numpy import genfromtxt
import operator
import json
import collections
import matplotlib.pyplot as plt
from matplotlib import cm
```

# Read Output of VeST

```python
# factor2 is related to movies
factor2 = genfromtxt('./result/MOVIE_FACTOR', delimiter='\t')
factor2 = np.delete(factor2, -1, 1)
```

# Read Meta-data related to movies (title, genres, etc.)

```python
movie_df = pd.read_csv('./result/movie_meta.csv')
sum_score = []
for i in range(len(factor2[0])):
    header = 'r'+str(i)
    movie_df[header] = factor2[:,i]
movie_df['sum_score'] = np.sum(factor2, axis=1)
```

```python
movie_df.head()
```

# Check the number of non-zero for each factor

```python
num_movies = len(factor2)
factor2_t = np.transpose(factor2)
```

```python
count_nonzero = []
for i in range(len(factor2_t)):
    count_nonzero.append(np.count_nonzero(factor2_t[i]))
```

```python
print(count_nonzero)
```

```python
# count the number of non-zero of factors for movies
nonzeros = np.count_nonzero(factor2, axis=1)
nonzero_df = pd.Series(nonzeros)
movie_df['nonzeros'] = nonzero_df
movie_df.head()
```

# Greatest Movies Discovery

```python
count = 0
r1_dic = {}
movies = []
for i in range(len(factor2)):
    if  np.count_nonzero(factor2[i]) == movie_df['nonzeros'].max():
        count += 1
        movies.append(i)

selected_df = movie_df[movie_df['movieId'].isin(movies)]
selected_df = selected_df.sort_values(by='sum_score', ascending=False)
print(len(selected_df))
selected_df["rank"] = selected_df["sum_score"].rank(ascending=False)
selected_df['rank']= selected_df['rank'].astype('int')
display(selected_df)
```

```python
show_df = selected_df[['title','rank']]
display(show_df)
```

# Draw heatmap of factor matrix for movies

```python
heatmap_df = movie_df.sort_values(by='nonzeros', ascending=False)
heatmap_df = heatmap_df.drop(heatmap_df[heatmap_df.nonzeros == 0].index)
heatmap_df.head()
```

```python
factor = [np.array(heatmap_df['r0']), np.array(heatmap_df['r1']), np.array(heatmap_df['r2']), 
                                    np.array(heatmap_df['r3']), np.array(heatmap_df['r4']), np.array(heatmap_df['r5'])]
factor = np.transpose(factor)
print(np.shape(factor))
print(factor[1])
```

```python
nfac = factor
r_dic=None
c_dic=None
a_min=None
a_max=None
aspect=None
prefix=''

print(prefix," fac shape", np.shape(nfac))
print("max: ", np.amax(nfac), " min: ", np.amin(nfac))

if a_min is None and a_max is None:
    a_max = round(np.std(nfac)*2.0,2)
    a_min = a_max*-1.0
    print("min max heatmap view set to", a_min, a_max)
if aspect is None :
    aspect = nfac.shape[1]/((nfac.shape[0]+1)*0.8)

if a_min is not None and a_max is not None:
    nfac = np.clip(nfac, a_min, a_max)

fig, ax = plt.subplots(figsize=(10,8))
cax = plt.imshow(nfac[:,:], cmap=cm.gist_heat_r,
        interpolation='nearest', aspect=aspect, vmin=0, vmax=a_max)

ax.set_title(prefix+' Movie Factors', fontsize=30)
if (r_dic is not None):
    ax.set_yticks(np.arange(0,len(r_dic)-1))
    ax.set_yticklabels(r_dic, fontsize=18)
if (c_dic is not None):
    ax.set_xticks(np.arange(0,len(r_dic)-1))
    ax.set_xticklabels(c_dic, fontsize=18, rotation=90)

ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Factors', fontsize=30)
ax.set_ylabel('Movies', fontsize=30)

cbar = fig.colorbar(cax, ticks=[a_min, a_min*0.5, 0, a_max*0.5, a_max])
cbar.ax.set_xticklabels([str(a_min), str(a_min*0.5), '0', str(a_max*0.5), str(a_max)])

plt.show()
# plt.savefig('{}.eps'.format(prefix+'heatmap_new'), bbox_inches='tight', format='eps', dpi=1000)
plt.close()
```
